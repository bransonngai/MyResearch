## First we need to read the data
The data in e.g. 20200904, which means the pervious day data ( coz we update the data in the morning)

```python
df = ba.hsbc.read_cbbc_xls('20200904', read_anyway= True)
df = df[df['Underlying Code'] == 'HSI'].groupby('Type')['Outstanding(M)'].sum()
BEAR_RATIO = round(df['Bear'] / (df['Bear'] + df['Bull'])* 100, 2)
BULL_RATIO = 100 - BEAR_RATIO

print(BEAR_RATIO, BULL_RATIO)
```

```python
# READ
def get(day):

    df = ba.hsbc.read_cbbc_xls(day, read_anyway= True)
    df = df[df['Underlying Code'] == 'HSI'].groupby('Type')['Outstanding(M)'].sum()
    BEAR_RATIO = round(df['Bear'] / (df['Bear'] + df['Bull'])* 100, 2)
    BULL_RATIO = 100 - BEAR_RATIO

    return BEAR_RATIO, BULL_RATIO
```

```python
date_range = ba.dk.date_ranger2('20190101', '20200904')
```

```python
result = []
for day in date_range:
    try:
        BEAR_RATIO, BULL_RATIO = get(day)
        result.append([day, BEAR_RATIO, BULL_RATIO])
    except Exception as e:
        print(day, e)
        
r = pd.DataFrame(result)
r.to_excel('mrlong.xls')
```
